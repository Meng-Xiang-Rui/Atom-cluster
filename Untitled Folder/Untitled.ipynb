```python

class TaskBase:
    """Base class for task"""

    def Dict_Init(self, data):
        pass

    def case_nan(self, *args):
        pass

    def Extract(self, *args):
        pass

    def Analysis(self, data):
        pass
```

```python
from datetime import datetime


class Time(TaskBase):
    def __init__(self):
        self.name = 'Time'

    def Dict_Init(self, data):
        data[self.name] = []

    def Extract(self, *args):
        file_path = args[0]
        data = args[1]
        time = datetime.strptime(
            ('20' + file_path[19:36]), '%Y.%m.%d-%H.%M.%S')
        data[self.name].append(time)

    def case_nan(self, *args):
        file_path = args[0]
        data = args[1]
        time = datetime.strptime(
            ('20' + file_path[19:36]), '%Y.%m.%d-%H.%M.%S')
        data[self.name].append(time)
```

```python
import numpy as np
import matplotlib.pyplot as plt


class Atom_number(TaskBase):
    def __init__(self):
        self.name = 'Atom_number'

    def Dict_Init(self, data):
        data[self.name] = []

    def case_nan(self, *args):
        number = np.nan
        data = args[1]
        data[self.name].append(number)

    def Extract(self, *args):
        allinone = args[2]
        number = np.array(allinone['FilledAtom1']).shape[0]
        data = args[1]
        data[self.name].append(number)

    def Analysis(data):
        time = []
        nAtom = np.array(data['Atom_number'])
        for i in data['Time']:
            time.append(str(i.strftime('%H:%M')))
        tick = [np.ceil(i) for i in np.linspace(0, len(time)-1, 5)]
        label = [time[int(i)] for i in tick]

        plt.figure(figsize=(12, 6))
        plt.plot(np.arange(len(time)), nAtom, 'o',
                 ms=8, label='Number of Atom')
        plt.xticks(tick, label)
        plt.grid(True)
        plt.title('Mean = {:.1f}, std = {:.0f}'.format(
            np.nanmean(nAtom), np.nanstd(nAtom)), fontsize=12)
        plt.xlabel('Time', fontsize=12)
        plt.ylabel('Number of Atoms', fontsize=12)
```

```python
import numpy as np
import matplotlib.pyplot as plt


class Atom_loc(TaskBase):
    def __init__(self):
        self.name = 'Atom_loc'

    def Dict_Init(self, data):
        data[self.name] = []

    def case_nan(self, *args):
        loc = [np.nan, np.nan]
        data = args[1]
        data[self.name].append(loc)

    def Extract(self, *args):
        allinone = args[2]
        loc = np.array(allinone['midPos1'][0])
        data = args[1]
        data[self.name].append(loc)

    def Analysis(data):
        time = []
        for i in data['Time']:
            time.append(str(i.strftime('%H:%M')))
        tick = [np.ceil(i) for i in np.linspace(0, len(time)-1, 5)]
        label = [time[int(i)] for i in tick]
        xc = np.array(data['Atom_loc'])[:, 1]
        yc = np.array(data['Atom_loc'])[:, 0]

        fig, ax1 = plt.subplots(figsize=(12, 6))
        ax2 = ax1.twinx()

        ax1.grid(True)

        p1, = ax1.plot(range(len(xc)), xc, 'o', ms=8, label='Xcenter')
        p2, = ax2.plot(range(len(xc)), yc, 'o', ms=8,
                       color='coral', label='Xcenter')

        ax1.xaxis.set_ticks(tick)
        ax1.xaxis.set_ticklabels(label)
        ax1.set_xlabel('Time', fontsize=12)
        ax1.set_ylabel('Xcenter', fontsize=12)
        ax2.set_ylabel('Ycenter', fontsize=12)
        ax1.yaxis.label.set_color(p1.get_color())
        ax2.yaxis.label.set_color(p2.get_color())
        ax1.tick_params(axis='y', colors=p1.get_color())
        ax2.tick_params(axis='y', colors=p2.get_color())

        lines = [p1, p2]
        ax1.legend(lines, [l.get_label() for l in lines])

        ax1.set_title(
            'Xcenter: Mean = {:.1f}, std = {:.0f}\nYcenter: Mean = {:.1f}, std = {:.0f}'
            .format(np.nanmean(xc), np.nanstd(xc), np.nanmean(yc), np.nanstd(yc)),
            fontsize=12)
        fig.tight_layout()
```

```python
import math
import numpy as np


def shift(data, distance):
    angle = np.array(math.radians(data['LatAng'][0]))
    base = np.array([np.array([math.cos(i), math.sin(i)]).T for i in angle])
    projection = np.dot(distance, base)
    a = data['LatGrid'][0]
    shift = [(projection[i] % a[i]) / a[i] * 2 * np.pi for i in range(len(a))]
    shift = [shift[i] if shift[i] <= np.pi else shift[i]-2*np.pi for i in range(len(a))]
    return shift
```

```python
from collections.abc import MutableSequence


class TaskList(TaskBase, MutableSequence):
    def __init__(self, tasks):
        self.name = 'TaskList'
        self.tasks = list(tasks)

    def __len__(self):
        return len(self.tasks)

    def __getitem__(self, key):
        return self.tasks[key]

    def __setitem__(self, key, value):
        if isinstance(value, TaskBase):
            self.tasks[key] = value
        else:
            raise TypeError(
                "value must be an instance of DataProcess.Tasks.TaskBase"
            )

    def __delitem__(self, index):
        del self.tasks[index]

    def __iter__(self):
        return iter(self.tasks)

    def __add__(self, other):
        return TaskList(self.tasks + other.tasks)

    def insert(self, index, value):
        if isinstance(value, TaskBase):
            self.tasks.insert(index, value)
        else:
            raise TypeError(
                "value must be an instance of DataProcess.Tasks.TaskBase"
            )

    def Dict_Init(self, data):
        for tb in self.tasks:
            tb.Dict_Init(data)

    def case_nan(self, *args):
        for tb in self.tasks:
            tb.case_nan(*args)

    def Extract(self, *args):
        for tb in self.tasks:
            tb.Extract(*args)

    def Analysis(self, data):
        for tb in self.tasks:
            tb.Analysis(data)
```

```python
import os

from scipy.io import loadmat


class process:
    
    def __init__(self, path, data = None, tasks = None):
        self.tasks = TaskList(tasks if tasks else [])
        self.tasks.append(Time())
        self.data = data if data else {}
        self.path = path

    def data_extract(self):
        self.tasks.Dict_Init(self.data)
        pic_list = os.listdir(path)[1:]
        for i in range(0, len(pic_list), 2):
            try:
                allinone_path = path + r'/analyResult/03.Reconstruct/params/' + pic_list[i][:-7] + r'-fig1/allInOne.mat'
                allinone = loadmat(allinone_path)
                number = np.array(allinone['FilledAtom1']).shape[0]
            except:
                self.tasks.case_nan(pic_list[i], self.data)
            else:
                self.tasks.Extract(pic_list[i], self.data, allinone)
    def pre_process(x):
        """nan => mean"""

        temp = x
        nan_list = []
        for i in range(len(temp)):
            if np.isnan(temp[i]):
                nan_list.append(i)
        mean = np.nanmean(temp)
        for i in nan_list:
            temp[i] = mean
        print(mean)
        return temp
```
